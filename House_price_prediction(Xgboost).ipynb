Importing independencies

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

Importing the california house price dataset


```python
house_price_dataset = sklearn.datasets.fetch_california_housing()
```

```python
house_price_dataset
```

Load the data into pandas data frame


```python
house_price_dataframe=pd.DataFrame(house_price_dataset.data,columns=house_price_dataset.feature_names)
```

```python
house_price_dataframe
```

```python
house_price_dataframe.shape
```

```python
house_price_dataframe.describe()
```

Adding the target (price) column to the dataframe

```python
house_price_dataframe['Price']=house_price_dataset.target
```

 The .target attribute is not part of the table (.data) â€” it's stored separately and needs to be added manually to the DataFrame.

```python
house_price_dataframe
```

```python
house_price_dataframe.shape
```

Check for missing values

```python
house_price_dataframe.isnull().sum()
```

```python
house_price_dataframe.describe()
```

Finding correlation between various features in the dataset

  1. positive correlation (one increaser second increases)
  2. negative correlation(one decreases second decreases)

```python
correlation=house_price_dataframe.corr()
```

1. cbar=color bar shown in the right
2. square = shape of the boxes
3. fmt= values after the decimal point
4. annot= feature names
5. annot_kws(annot sie): basically the font size.
6. cmap=color of the map: blueS, redS

```python
#constructing a heatmap to understand the correlation
plt.figure(figsize=(6,6))
sns.heatmap(correlation,cbar=True,square=True,fmt='.1f',annot=True,
            annot_kws={'size':8},cmap='coolwarm')
```

splitting data into features and lables


```python
x=house_price_dataframe.drop(columns="Price", axis=1)
y=house_price_dataframe['Price']
```

```python
print(x)
print(y)
```

Splitting the data into tarin and test

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

Model tarining

XG boost regressor

```python
#loading the model
model=XGBRegressor()
```

```python
#trainig the model with xtrain
model.fit(x_train,y_train)
```

prediction on parallel data

```python
prediction=model.predict(x_train)
print(prediction)
```

Compare the predicted value with the actual value or the target values that is y train

```python
#R quare error
score_1= metrics.r2_score(y_train,prediction)
print("R squared error=:", score_1)  #if value close to 1 means our model is working well

#Mean absolute error
score_2=metrics.mean_absolute_error(y_train,prediction)
print("Mean absolute error=", score_2) #if value close to 0 means our model is working well
```

Visualize the actual prices and predicted prices

```python
# Visualize the actual prices and predicted prices for the training data
plt.scatter(y_train, prediction)
plt.xlabel("Actual prices")
plt.ylabel("Predicted prices")
plt.title("Actual prices vs Predicted prices")
plt.show()
```

Prediction on test data

```python
prediction=model.predict(x_test)
print(prediction)
```

```python
#R quare error
score_1= metrics.r2_score(y_test,prediction)
print("R squared error=:", score_1)  #if value close to 1 means our model is working well

#Mean absolute error
score_2=metrics.mean_absolute_error(y_test,prediction)
print("Mean absolute error=", score_2) #if value close to 0 means our model is working well
```
